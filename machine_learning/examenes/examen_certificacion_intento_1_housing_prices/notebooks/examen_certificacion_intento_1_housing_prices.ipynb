# Examen de certificación – Intento 1

<strong>Objetivo:</strong> Este examen tiene como objetivo evaluar la capacidad del estudiante para limpiar, explorar, implementar y evaluar modelos de regresión en un dataset de precios de casas. Utilizaremos el «[House Prices Dataset](https://www.kaggle.com/competitions/house-prices-advanced-regression-techniques/data)» disponible en Kaggle. Los estudiantes deben demostrar habilidades prácticas en la manipulación de datos, creación de visualizaciones y modelado predictivo.

## Contexto y Descripción del Dataset

El «House Prices Dataset» contiene datos sobre diversas características de casas residenciales en Ames, Iowa. La tarea es predecir el precio final de cada casa (variable objetivo) en función de sus características.

## Requisitos

### 1. Limpieza de Datos:

* <strong>Identificación y eliminación de valores duplicados:</strong> Asegúrate de que no haya registros duplicados que puedan sesgar los resultados del análisis.
* <strong>Verificación y ajuste de tipos de datos:</strong> Verifica que cada columna tenga el tipo de dato correcto (numérico o categórico) y ajusta si es necesario.
* <strong>Corrección de inconsistencias en valores categóricos:</strong> Revisa las categorías de las variables y unifica aquellos valores que puedan estar escritos de diferentes maneras pero que representen lo mismo.
* <strong>Manejo de valores faltantes adecuadamente:</strong> Identifica y maneja los valores faltantes utilizando técnicas apropiadas como la imputación de la mediana, media o moda, según corresponda.

### 2. Exploración de Datos:

* <strong>Visualizaciones univariadas y multivariadas:</strong> Crea histogramas, gráficos de barras, diagramas de dispersión y mapas de calor para entender la distribución y las relaciones entre las variables.
* <strong>Estadísticas descriptivas:</strong> Calcula medidas de tendencia central (media, mediana, moda) y de dispersión (rango, desviación estándar) para cada característica del dataset.

### 3. Implementación de Modelos:

* <strong>Modelos de Regresión:</strong> Implementa modelos de Linear Regression y LightGBM (LGBM).
* <strong>Evaluación de Modelos:</strong> Evalúa los modelos utilizando métricas como MSE, RMSE, y R^2.
* <strong>Comparación de Rendimiento:</strong> Compara los resultados de ambos modelos y discute cuál es el más adecuado para este dataset.

## Entrega

## Consideraciones Éticas y Tecnológicas

### Consideraciones Éticas:

* <strong>Transparencia y Reproducibilidad:</strong> Asegúrate de que todos los pasos del análisis sean claros y reproducibles. Otros investigadores deben poder seguir tus pasos y llegar a los mismos resultados.
* <strong>Imparcialidad y Sesgo:</strong> Revisa si existen sesgos en los datos que puedan afectar la imparcialidad del modelo. Es importante que los modelos no discriminen injustamente entre diferentes grupos de datos.

### Consideraciones Tecnológicas:

* <strong>Herramientas Utilizadas:</strong> Utiliza herramientas estándar como Python, Jupyter Notebook, Pandas, Scikit-learn, Matplotlib y Seaborn.
* <strong>Escalabilidad:</strong> Considera cómo las técnicas aplicadas podrían escalarse para manejar conjuntos de datos más grandes y complejos.
* <strong>Optimización de Modelos:</strong> Aunque este examen no se enfoca en la optimización de hiperparámetros, se debe tener en cuenta para futuras implementaciones y mejorar el rendimiento de los modelos.
