### Limpieza de datos

```python
# Primero, chequeamos si hay o no filas duplicadas.
cant_duplicados = len(df[df.duplicated()])
print(f"Cantidad de filas duplicadas: {cant_duplicados}")
```

Como no existen filas duplicadas, no es necesario eliminar nada.

En caso de que existan duplicados, pueden ser eliminadas as칤: df.drop_duplicates(keep='first', inplace=True)

Notar que keep determina cual versi칩n es la que va a quedar

```python
# Adicionalmente, se pueden identificar si existen datos duplicados s칩lo en las columnas Location y Date_Time
cant_duplicados_location_datetime = len(df[df.duplicated(["Location", "Date_Time"])])
print(f"Cantidad de filas por ubicaci칩n y fecha: {cant_duplicados}")
```

Como no existen filas duplicadas, no es necesario eliminar nada.

En caso de que existan duplicados, pueden ser eliminadas as칤: df.drop_duplicates(["Location", "Date_Time"], keep='first', inplace=True)

Adicionalmente, en caso de que hayan duplicados, hay que ver cu치les valores eliminar y cuales no, ya definiendo un criterio de acuerdo a los datos.

```python
# Para mayor facilidad, es posible convertir la columna Date_time que es de tipo cadena a datetime de python.
df["Date_Time"] = pd.to_datetime(df["Date_Time"], format="%Y-%m-%d %H:%M:%S")

df.info()
df.head()
```

```python
# Volvemos a mostrar las estad칤sticas de los datos.
df.describe()
```

### Transformaci칩n de Datos

```python
# Crea nuevas columnas basadas en operaciones con las columnas existentes

# A modo de prueba, se agrega una nueva columna Temperature_F convirtiendo la temperatura en Celsius a Fahrenheit
df["Temperature_F"] = (df["Temperature_C"] * (9/5)) + 32

df.head()
```

```python
# Tambi칠n es posible hacerlo con .apply()

df["Temperature_F"] = df["Temperature_C"].apply(lambda x: (x * 9/5) + 32)

df.head()
```

```python
# A modo de pruebas, tambi칠n se normaliza la columna Humidity_pct a valores entre 0 y 1.

df["Humidity_pct"] = df["Humidity_pct"] / 100

# Tambi칠n se puede: df["Humidity_pct"] = df["Humidity_pct"].apply(lambda x: x / 100)

df.head()
```

### An치lisis de Datos

```python
# Para realizar un ejemplo de agrupaci칩n y de funciones de agregaci칩n, una idea ser칤a
# obtener los promedios de temperatura en Celsius y humedad en porcentaje en cada ciudad
# y en un rango de tiempo. Tambi칠n se podr칤an visualizar la cant de datos por ciudad.
from datetime import datetime, timedelta

start_time = datetime(year=2024, month=1, day=1)
end_time = start_time + timedelta(days=3 * 30)

df_en_fechas = df[(df["Date_Time"] >= start_time) & (df["Date_Time"] <= end_time)]

df_en_fechas.groupby("Location")[["Temperature_C", "Humidity_pct"]].agg(["std", "mean", "size", "min", "max", "var"])
```

```python
# Otro rango de fechas

start_time = datetime(year=2024, month=4, day=1)
end_time = start_time + timedelta(days=3 * 30)

df_en_fechas = df[(df["Date_Time"] >= start_time) & (df["Date_Time"] <= end_time)]

df_en_fechas.groupby("Location")[["Temperature_C", "Humidity_pct"]].agg(["std", "mean", "size", "min", "max", "var"])
```

```python
# Tambi칠n, se puede agregar una columna que indique el mes y el a침o,
# para luego agrupar por dicha columna y realizar m치s estad칤sticas.
df["Year_Month"] = df["Date_Time"].apply(lambda dt: dt.strftime("%Y-%m"))
```

```python
# Ordenamos por ciudad y luego por fecha para mayor facilidad
# a la hora de visualizar los datos.
df.sort_values(by=["Location", "Date_Time"], inplace=True)
df.head(31)
```

```python
# Si quisieramos por ej visualizar datos de temperatura
# para San Diego en Enero 2024:

df[(df["Location"] == "San Diego") & (df["Year_Month"] == "2024-01")]
```

Al visualizar r치pidamente los datos ya ordenados, es posible notar un patr칩n extra침o en los datos de temperatura, debido a los saltos muy abruptos. Esta situaci칩n es extra침a si consideramos que las mediciones de temperatura no tienen cambios bruscos tan r치pidamente.
